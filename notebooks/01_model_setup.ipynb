## 3. Pull Available Models

````python
import subprocess

# Available models (no API keys needed)
models = [
    "llama3.1",      # 8B parameters - Recommended
    "llama3.1:70b",  # 70B parameters - Better quality, needs more RAM
    "mistral",       # 7B parameters - Fast and efficient
    "codellama",     # Code-focused model
    "phi3",          # 3.8B parameters - Lightweight
]

# Pull the recommended model
print("Pulling llama3.1 model...")
subprocess.run(["ollama", "pull", "llama3.1"])
````

## 4. Test Model Connection

````python
from langchain_community.llms import Ollama

# Initialize local model (no API key needed)
llm = Ollama(
    base_url="http://localhost:11434",
    model="llama3.1"
)

# Test the model
response = llm.invoke("What is workflow automation?")
print(response)
````

## 5. Test Workflow Planning

````python
from langchain.prompts import PromptTemplate

# Create a planning prompt
prompt = PromptTemplate(
    input_variables=["task"],
    template="""You are a workflow planning agent. Break down this task into steps:

Task: {task}

Provide a numbered list of steps:"""
)

# Test planning
chain = prompt | llm
result = chain.invoke({"task": "Onboard a new client with Jira project and Stripe billing"})
print(result)
````

## 6. Model Performance Comparison

````python
import time

def test_model_performance(model_name):
    """Test model response time and quality"""
    llm = Ollama(base_url="http://localhost:11434", model=model_name)
    
    start = time.time()
    response = llm.invoke("List 3 steps to create a Jira project")
    elapsed = time.time() - start
    
    return {
        "model": model_name,
        "time": f"{elapsed:.2f}s",
        "response_length": len(response)
    }

# Test available models
for model in ["llama3.1", "mistral", "phi3"]:
    try:
        result = test_model_performance(model)
        print(f"\n{result['model']}: {result['time']} ({result['response_length']} chars)")
    except Exception as e:
        print(f"\n{model}: Not available (pull with: ollama pull {model})")
````
